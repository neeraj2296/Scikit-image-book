```python
from scipy import ndimage, misc
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(12,5))
ax1 = fig.add_subplot(121)  # left side
ax2 = fig.add_subplot(122)  # right side

ascent = misc.ascent()
result = ndimage.zoom(ascent, 3.0)

ax1.imshow(ascent)
ax2.imshow(result)

plt.show()
```
